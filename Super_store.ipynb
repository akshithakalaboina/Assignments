## Outliers

### Satistical method

```python
D = df.select_dtypes(exclude=["object", 'datetime64[ns]'])
```

```python
for i in D.columns:
    q1 = D[i].quantile(0.25)
    q3 = D[i].quantile(0.75)
    IQR = q3 - q1
    lower = q1 -(1.5 * IQR)
    upper = q3 +(1.5 * IQR)
    print(f"{i} lower limit is {lower} ")
    print(f'{i} upper limit is {upper} ')
```
