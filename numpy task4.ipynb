```python
import numpy as np
```

```python
iris = np.genfromtxt(r"C:\Users\hi\Downloads\Iris_clean_-_Copy.csv",dtype = 'float16',delimiter = ",")
```

```python
iris
```

```python
sepal_width= iris[1:,1]
sepal_width
```

```python
np.nanmedian(sepal_width) #median
```

```python
sepal_width = np.where(np.isnan(sepal_width),3,sepal_width)  #missing values
sepal_width
```

```python
outlier_sepal_width = len(sepal_width[sepal_width>6]) #no.of outliers in sepal width
outlier_sepal_width
```

```python
final_sepal_width = np.where(sepal_width>6,4,sepal_width)
final_sepal_width
```
